### Lmplot represents a positve slope bw Temp & Windspeed

```python
# bw windspeed & Avg. Temperature
df=pd.read_csv("AQI_data.csv")
sns.lmplot(x='T', y='V', data=df, scatter_kws={'s': 20, 'color': 'darkblue'}, line_kws={'color': 'red'})
plt.title("Temp(Celcisus) vs Windspeed(km/h)")
```

### LMPlot with positive slope bw Atmospheric Pressure & PM 2.5

```python
sns.lmplot(x='SLP', y='PM 2.5', data=df, scatter_kws={'s': 20, 'color': 'darkblue'}, line_kws={'color': 'red'})
plt.title("Atmospheric Pressure(hPa) vs PM 2.5")
```
